```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

import pandas as pd
from sklearn.model_selection import train_test_split
from scipy import stats
```

```python
#EasyPark data
data = pd.read_csv('easypark_data.csv')
data = data[data.columns.values[2:]]

data.head()
```

```python
X = data[data.columns.values[:-3]].values
expert_Y = data[data.columns.values[-3]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, expert_Y, test_size=0.3, random_state=42)
```

```python
reg = RandomForestRegressor()
reg.fit(X_train, y_train)
pred = reg.predict(X_test)
stats.ttest_rel(pred, y_test).pvalue
```
